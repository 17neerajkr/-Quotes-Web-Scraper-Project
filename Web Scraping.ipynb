```python
for q in soup.find_all("span", class_="text"):
    text = q.text.strip()   # clean whitespace and \n
    if text.startswith("A"):    # check starts with A
        print(text)
```

```python
for q in soup.find_all("span", class_="text"):
    text = q.text.strip()
    clean = text.lstrip("“”\"'")   # remove quote marks if present
    if clean.startswith("A"):
        print(text)
```
